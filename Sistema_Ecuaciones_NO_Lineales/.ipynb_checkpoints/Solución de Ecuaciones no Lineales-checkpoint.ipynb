# Introducción

Para hallar la solución de ecuaciones no lineales, se usa métodos basados en aproximaciones numéricas. Se usan métodos iterativos para resolver la ecuación $f(x)=0$.

# Método de Bisección

```octave
%Metodo de Biseccion
%Ejemplo

%Puntos a y b iniciales
a = 0.5;
b = 1.5;
%Variable auxiliar para el contador
aux=0;
%Funcion de ejemplo
f=@(w ) ((9.8*(sinh(w)-sin(w)))/(2*w.^2)-1);


#El bucle while corre mientras el valor absoluto de (b-a) sea mayor a la tolerancia 
while abs(b-a)>0.00001
#Comprueba si hay un cero entre a y b
if f(a)*f(b)<0
# c es el valor medio entre a y b
c=(a+b)/2 ;
# si cumple que entre a y c hay un cero, c sera el nuevo b
if f(a)*f(c)<0
b=c;
#Caso contrario c sera el nuevo a
else
a=c;
end
%Se aumenta uno a contador
aux=aux+1;
else
disp ('Error en intervalo') ;
return
endif
endwhile
#El ultimo valor de c al que converge
c
#Cuantas iteraciones fueron necesarias
aux
```

# Método de la Secante

```octave
%Ejercicio Metodo de la Secante

syms x
f(x)=2+2*x^3+sin(x);
% Para obterner la funcion en LaTeX latex(f)
x0=-11.2
x1=10.7

x2=x1-((x1-x0)*eval(f(x1))/(eval(f(x1))-eval(f(x0))));

while abs(x2-x1)>0.00001
  x0=x1;
  x1=x2;
x2=x1-((x1-x0)*eval(f(x1))/(eval(f(x1))-eval(f(x0))));
endwhile
resul=x2
```

# Método de Newton

```octave
%Ejercicio Metodo de Newton

syms x
f(x)=2+2*x^3+sin(x)
x0=-10;
g(x)=diff(f,x)

%Formula Newton

x1=x0-eval(f(x0))/eval(g(x0));

while abs(x1-x0)>0.00001
x0=x1;
x1=x0-eval(f(x0))/eval(g(x0));
endwhile

resul=x1
```
